```python
# Importando as bibliotecas necessárias da Prova do modulo 06
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Carregando o dataset do Titanic
df = sns.load_dataset('titanic')

# Exibindo os primeiros registros do dataframe
print(df.head())

# Tabela de contingência entre 'sex' e 'survived'
contingency_table = pd.crosstab(df['sex'], df['survived'])
print(contingency_table)

# Calculando o teste de Qui-Quadrado
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Qui-Quadrado: {chi2}, p-valor: {p}")

# Criando um gráfico para visualizar a relação
sns.countplot(data=df, x='sex', hue='survived')
plt.title('Relação entre Sexo e Sobrevivência')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando os valores de x1 e x2
x1 = np.linspace(-10, 10, 400)
x2 = np.linspace(-10, 10, 400)

X1, X2 = np.meshgrid(x1, x2)

# Definindo a função f(x1, x2)
f = X1**2 + X2**2

# Criando o gráfico de contorno
plt.contour(X1, X2, f, levels=[9, 10, 11], colors='blue')

# Adicionando um círculo branco nos pontos onde f(x1, x2) está próximo de 10
circle = plt.Circle((0, 0), np.sqrt(10), color='white', fill=False, linewidth=3)
plt.gca().add_artist(circle)

# Configurações do gráfico
plt.title("Região onde x1² + x2² ≈ 10")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando o dataframe
df = pd.read_csv("dadosgovbr---2014.csv", sep=';', encoding="latin-1")

# Pergunta 1: Número de reclamações por Estado
reclamacoes_estado = df['UF'].value_counts()
print(reclamacoes_estado)

# Plotando o gráfico das reclamações por Estado
reclamacoes_estado.plot(kind='bar')
plt.title('Número de Reclamações por Estado')
plt.xlabel('Estado')
plt.ylabel('Número de Reclamações')
plt.show()

# Pergunta 2: Quantidade de reclamações por sexo
reclamacoes_sexo = df['Sexo Consumidor'].value_counts()
print(reclamacoes_sexo)

# Plotando o gráfico de reclamações por sexo
reclamacoes_sexo.plot(kind='bar')
plt.title('Reclamações por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Número de Reclamações')
plt.show()

# Pergunta 3: Correlação entre tempo de resposta e nota do consumidor
# (Considera que existem colunas de "tempo de resposta" e "nota do consumidor")
corr = df['Tempo Resposta'].corr(df['Nota do Consumidor'])
print(f"Correlação entre Tempo de Resposta e Nota do Consumidor: {corr}")

# Pergunta 4: Proporção de reclamações não respondidas
reclamacoes_nao_respondidas = df[df['Respondida'] == 'Não'].shape[0]
total_reclamacoes = df.shape[0]
proporcao_nao_respondidas = reclamacoes_nao_respondidas / total_reclamacoes
print(f"Proporção de Reclamações Não Respondidas: {proporcao_nao_respondidas:.2%}")

# Pergunta 5: Outras possíveis análises

# Análise 1: Qual é o tempo médio de resposta por estado?
tempo_resposta_estado = df.groupby('UF')['Tempo Resposta'].mean().sort_values()
tempo_resposta_estado.plot(kind='bar')
plt.title('Tempo Médio de Resposta por Estado')
plt.xlabel('Estado')
plt.ylabel('Tempo Médio de Resposta')
plt.show()

# Análise 2: Existe correlação entre a idade do consumidor e a nota dada?
# (Considera-se que existe uma coluna de "Idade Consumidor")
corr_idade_nota = df['Idade Consumidor'].corr(df['Nota do Consumidor'])
print(f"Correlação entre Idade do Consumidor e Nota: {corr_idade_nota}")
```
