```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o dataset
df = pd.read_csv("dataframe_exercicio_modulo_9.csv")

# Verificar as primeiras linhas do dataset
print(df.head())

# Informações gerais sobre o dataset
print(df.info())

# Estatísticas descritivas das variáveis numéricas
print(df.describe())

# Verificar a distribuição da variável target (G3)
plt.figure(figsize=(10, 6))
sns.histplot(df['G3'], kde=True, bins=20)
plt.title("Distribuição da Nota Final (G3)")
plt.xlabel("G3")
plt.ylabel("Frequência")
plt.show()

# Verificar correlação entre variáveis numéricas
plt.figure(figsize=(12, 8))
# Convert columns to numeric type, errors='coerce' will replace invalid values with NaN
numeric_df = df.apply(pd.to_numeric, errors='coerce')
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Matriz de Correlação")
plt.show()

# Contagem de variáveis categóricas
categorical_columns = ['sex', 'famsize', 'Pstatus', 'schoolsup', 'famsup', 'paid', 'activities', 'internet', 'romantic']
for column in categorical_columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=column, data=df)
    plt.title(f"Distribuição de {column}")
    plt.show()
```
